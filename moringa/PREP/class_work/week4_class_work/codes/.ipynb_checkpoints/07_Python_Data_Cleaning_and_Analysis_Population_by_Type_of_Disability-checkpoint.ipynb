## 1.4 Cleaning our Dataset

Let us perform data cleaning procedures below providing a documentation for our actions and reasons. We will perform as many data cleaning procedures as we think suitable for the various dimensions of data.

### 1.41) Validity: <font color="green">Challenges</font>

```python
# Procedure 1: Irrelevant Data Observation
# Data Cleaning Action: Dropping Location_1 attribute
# Explanation: We won't need it during Analysis. No question to be answered requires that column.
#
frame.drop('Location_1', inplace=True, axis=1)
frame
```

```python
# Procedure 2: Irrelevant Data Observation
# Data Cleaning Action: Drop OBJECTID attribute
# Explanation: Its irrelevant to our table i.e. attribute starts from 0 and ends of 48 i.e. No meaning at all.
# 
frame.drop('OBJECTID', inplace=True, axis=1)
frame
```

```python
# Procedure 3: white spaces observation
# Data Cleaning Action: drop all the white spaces that might exist
# Explanation: We drop white spaces to attain uniformity in our data
#
for column in frame.columns:
    try:
        frame[column] = frame[column].str.strip()
    except:
        pass
frame
```

```python
# Procedure 4: formatting issue
# Data Cleaning Action: Drop % sign in percentage values and change to actuals
# Explanation: This allows us to work with the data in our computations since the percentages is not a clear representation of 
#numbers
frame_actual = pd.DataFrame()

for column in frame.columns:
    try:
        frame_actual[column] = [int(x) for x in frame[column].str.rstrip('%').astype('float')/100.0 * frame['Total_Count']]
    except:
        pass
frame_actual['Total_Count'] = [int(x) for x in frame['Total_Count']]
frame_actual
```

### 1.42) Accuracy <font color="green">Challenges</font>

```python
# Procedure 1: Variation of summationing
# Data Cleaning Action: Make a column representing the total_number of cases reported
# Explanation: Total_count and total_cases differs since one can have more that a single state of disability
#
#frame['Total_Percentages']
count = 0
total_cases = []
while count < frame.shape[0]:
    region_total = int(frame_actual.iloc[count, 1:9].sum())
    total_cases.append(region_total)
    count += 1

frame_actual['Total_Cases'] = total_cases
frame_actual['County'] = frame.County
frame_actual
```

## 1.5 Answering Questions

Let's answer the following questions with our dataset using python.

```python
# Challenge 1
# Which county had the highest no. of registered deaf persons?
# 
max_deaf = frame_actual['Deaf'][0:-1].max()
result = frame_actual[frame_actual['Deaf']==max_deaf]
result.County
```

```python
# Challenge 2
# Which county had the highest no. of registered persons with a missing hand?
#
Mini = frame_actual['Missing_Hand'][0:-1].max()
result = frame_actual[frame_actual['Missing_Hand']==Mini]
result.County
```

```python
# Challenge 3
# Which county had the highest no. of registered persons with a missing foot?
# 
Mini = frame_actual['Missing_Foot'][0:-1].max()
result = frame_actual[frame_actual['Missing_Foot']==Mini]
result.County
```

```python
# Challenge 4
# Which county had the highest no. of registered lame persons?
# 
Mini = frame_actual['Lame'][0:-1].max()
result = frame_actual[frame_actual['Lame']==Mini]
result.County
```

```python
# Challenge 5
# Which county had the lowest no. of registered blind persons?
# 
mini = frame_actual['Blind'][0:-1].min()
result = frame_actual[frame_actual['Blind']==mini]
result.County
```

```python
# Challenge 6
# Which county had the highest third no. of registered deaf persons?
# 

result = frame_actual.sort_values(by=['Deaf'], ascending=False)
result.iloc[3, -1]
```

```python
# Challenge 7
# In descending order, which top 5 counties had the highest no. of registered dumb persons?
# 
result = frame_actual.sort_values(by=['Dumb'], ascending=False)
counties = result.iloc[1:6, :]
counties.County
```

```python
# Challenge 8
# In ascending order, which top 5 counties had the highest no. of registered persons with a mental disability? 
# 
result = frame_actual.sort_values(by=['Mental'], ascending=False)
counties = result.iloc[1:6, :].sort_values(by=['Mental'], ascending=True)
counties.County
```

```python
# Challenge 9
# Which counties had no registerd blind persons nor deaf persons?
# 
result = frame_actual[(frame_actual['Blind']==0) & (frame_actual['Deaf']==0)]
result.County
```

```python
# Challenge 10
# Which disability was the most registered across all the counties?
# 
col_use = frame_actual.iloc[:-1, 0:8]
value = col_use.sum()
maxi = value.max()

final = {}
count = 0
for key in value.keys():
    val = value[count]
    final[key] = val
    count = count + 1

for key, value in final.items():
    if value == maxi:
        print(key)
```

```python
# Challenge 11
# Which disability was the least registered across all the counties?
#
col_checking = frame_actual.iloc[:-1, 0:8]
value = col_checking.sum()
maxi = value.min()

final = {}
count = 0
for key in value.keys():
    val = value[count]
    final[key] = val
    count = count + 1

for key, value in final.items():
    if value == maxi:
        print(key)
```

```python
# Challenge 12
# What was the average no. of registered persons with a disability?
#
total_dis = pd.DataFrame()
total_dis['County'] = frame_actual['County']
total_dis['total'] = frame_actual['Missing_Hand'] + frame_actual['Missing_Foot'] + frame_actual['Lame'] + frame_actual['Blind'] + frame_actual['Deaf'] + frame_actual['Dumb'] +frame_actual['Mental']+frame_actual['Paralyzed'] + frame_actual['Other']

print(f"Average of disabled across all counties is {total_dis.total.mean()}")
```

```python
# Challenge 13
# Which three counties had least registered persons with disabilities?
#
result = total_dis.sort_values(by='total')
result.iloc[:3, :]
```

```python
# Challenge 14
# What was the total no of registered persons with a disability across all counties?
#
result.total.sum()
```
