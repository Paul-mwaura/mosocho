```
import csv
import pandas as pd
```

```
#read csv as pandas frame
with open('/content/Uganda_Karamoja_Subcounty_Crop_Yield_Population.csv', 'r') as f:
  uganda = pd.read_csv(f, index_col=0, encoding='utf-8')

#check for existance then create a new one if one exists
%sql DROP TABLE If EXISTS uganda
%sql Persist uganda
```

```
#dispaly table structure
%sql PRAGMA table_info(uganda)
```

```
#QUESTION 1
#Displaying a list of Sub Counties, Population and Area
%%sql

SELECT SUBCOUNTY_NAME, POP, Area FROM uganda;
```

```
#QUESTION 2
#Sort the list of districts by total crop area (descending order)
%%sql

SELECT DISTRICT_NAME FROM uganda
ORDER BY Crop_Area_Ha DESC;
```

```
#QUESTION 3
#Select only the Sub counties from the Moroto district, order them alphabetically and show their production of sorghum.

%%sql

SELECT SUBCOUNTY_NAME, S_Prod_Tot FROM uganda
WHERE DISTRICT_NAME LIKE '%oroto%'
ORDER BY SUBCOUNTY_NAME ASC;
```

```
#QUESTION 4
#Compute the total Maize production per District.

%%sql

SELECT DISTRICT_NAME, COUNT(SUBCOUNTY_NAME), SUM(M_Prod_Tot) FROM uganda
GROUP BY DISTRICT_NAME;
```

```
#QUESTION 5
#Compute the number of Sub counties where Maize is produced and the total Maize production per District.

%%sql

      --number of counties producing Maize--
SELECT COUNT(SUBCOUNTY_NAME) FROM uganda
WHERE M_Yield_Ha IS NOT NULL;
```

```
#QUESTION 6
#Compute the overall Crop area in all Sub counties where population is over 20000.

%%sql

SELECT SUM(Crop_Area_Ha) FROM uganda
WHERE POP > 20000;
```

```
#QUESTION 7A
#Sort the Maize production in descending order by Districts, only taking into account Sub counties where Maize area is larger than 
#Sorghum area, display the number of Sub counties per district matching that criteria

%%sql

SELECT DISTRICT_NAME, COUNT(SUBCOUNTY_NAME) as `TOTAL SUBCOUNTIES`, SUM(M_Prod_Tot) AS `TOTAL PRODUCTION` FROM uganda
WHERE S_Area_Ha < M_Area_Ha
GROUP BY DISTRICT_NAME
ORDER BY DISTRICT_NAME DESC
;
```

```
#alternative question 7 without displaying the number of sub counties
%%sql

SELECT DISTRICT_NAME, SUBCOUNTY_NAME, M_Prod_Tot FROM uganda
WHERE S_Area_Ha < M_Area_Ha
ORDER BY DISTRICT_NAME DESC
;
```
